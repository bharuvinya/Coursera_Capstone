<b>Extracting the Latitude and Longitude of Canada Ontario</b>

```python
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner
link = "http://cocl.us/Geospatial_data"
df1 = pd.read_csv(link)
df1.head()
```

```python
df1.shape
```

<b>Changing the column name Postal code to Postcode to merge the two data frames together:</b>

```python
df1.columns = ['Postcode','Latitude','Longitude']

cols = df1.columns.tolist()
cols
```

<b>Read in the CSV file saved in the previous assignment:</b>

```python
link = "https://raw.githubusercontent.com/Shekhar-rv/Coursera_Capstone/master/df_can.csv"
df = pd.read_csv(link,index_col=0)
df.head()
```

<b>Merging the two data frames together based on their Postcode:</b>

```python
df_new = pd.merge(df, df1, on='Postcode')
df_new.head()
```
